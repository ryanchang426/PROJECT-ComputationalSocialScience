```python
s = Service("C:\\Users\\ThinkPad\\AppData\\Local\\Google\\Chrome\\Application\\chromedriver")
driver = webdriver.Chrome(service=s)
driver.maximize_window()
driver.get('https://flights.ctrip.com/online/list/oneway-bjs-sha?depdate=2022-04-20&cabin=y_s_c_f&adult=1&child=0&infant=0')

for i in range(4):
#     browser.execute_script("window.scrollBy(0, 1000)")
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)
time.sleep(1)

elements_flight_part = driver.find_elements(by=By.CLASS_NAME, value='flight-part')
elements_flight_item = driver.find_elements(by=By.CLASS_NAME, value='flight-item.domestic')

html_elements_flight_part = elements_flight_part[0].get_attribute("outerHTML")
my_elements_flight_part = BeautifulSoup(html_elements_flight_part, 'html.parser')

my_elements =  [0 for i in range(len(elements_flight_item))] 
soup_box =  [0 for i in range(len(elements_flight_item))] 
for i in range(len(elements_flight_item)):
    my_elements[i] = elements_flight_item[i].get_attribute("outerHTML")
    html_doc = my_elements[i]
    soup_box[i] = BeautifulSoup(html_doc, 'html.parser')

print(len(my_elements))

# driver.close()
```

```python
print(my_elements_flight_part.prettify())
print('--------------------------------------------------------------------------------')

for i in range(len(my_elements)):

    print('第',i+1,'份')
    print(soup_box[i].prettify())
    print('--------------------------------------------------------------------------------')
```

```python
index = [0 for i in range(len(soup_box))]
price = [0 for i in range(len(soup_box))]


for i in range(len(soup_box)):
    soup = soup_box[i]
    index[i] = i+1
    price[i] = soup.findAll(name="span", attrs={"class" :"price"})[0].text[1:]
    unit = soup.findAll(name="dfn")[0].text
    price[i] = price[i] + unit
    
    print(index[i], price[i])
```

```python
depart_place = [0 for i in range(len(soup_box))]
depart_time = [0 for i in range(len(soup_box))]

for i in range(len(soup_box)):
    soup = soup_box[i]
    
    depart = soup.findAll(name="div", attrs={"class" : "depart-box" })[0]
    depart_place[i] = depart.findAll(name="span", attrs={"id" : re.compile('departureFlightTrain') })[0].text 
    depart_time[i] = depart.findAll(name="div", attrs={"class" : "time" })[0].text
    
    print(depart_place[i], depart_time[i])
```

```python
arrive_place = [0 for i in range(len(soup_box))]
arrive_time = [0 for i in range(len(soup_box))]

for i in range(len(soup_box)):
    soup = soup_box[i]
    
    arrive = soup.findAll(name="div", attrs={"class" : "arrive-box" })[0]
    arrive_place[i] = arrive.findAll(name="span", attrs={"id" : re.compile('arrivalFlightTrain') })[0].text 
    arrive_time[i] = arrive.findAll(name="div", attrs={"class" : "time" })[0].text
    
    print(arrive_place[i], arrive_time[i])
```

```python
table = xlwt.Workbook(encoding='utf-8',style_compression=0)

sheet = table.add_sheet('plane price',cell_overwrite_ok=True)

col = ('index','price','depart_place','depart_time','arrive_place','arrive_time','appendix')


for i in range(0,len(col)):
    sheet.write(0,i,col[i])

for i in range(0,len(soup_box)):
    sheet.write(i+2, 0, index[i])
    sheet.write(i+2, 1, price[i])
    sheet.write(i+2, 2, depart_place[i])
    sheet.write(i+2, 3, depart_time[i])
    sheet.write(i+2, 4, arrive_place[i])
    sheet.write(i+2, 5, arrive_time[i])

# my_elements_flight_part.text
sheet.write(1, 6, my_elements_flight_part.text)

savepath = r'C:\Users\ThinkPad\jupyter_notebook\数据挖掘\计算社会科学\携程机票爬虫' + '\plane_price.xls'
table.save(savepath)  

pd.read_excel(savepath)
```
